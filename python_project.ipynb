Basic chat boot

```python
def chatbot():
    print("Chatbot:Hi I'm Your Chatboot. Type 'bye to exit.\n")

    while True:
        user_input = input("You: ").lower()

        if user_input in ["hi","hello","hey"]:
            print("Chatbot: Hi there!")

        elif user_input in ["how are you"]:
            print("Chatbot: I'm fine, thanks! How about you?")

        elif user_input in["i amfine","fine","good"]:
            print("Chatbot: Great to hear that!")

        elif user_input in ["bye","exit","quit"]:
          print("chatbot: Goodbye! Have a great day!")
          break
        else:
            print("Chatbot: Sorry, I don't understand that. Try saying 'hello' or 'bye'.")

# Run the chatbot
chatbot()
```

Stock Portfolio Tracker

```python
stock_prices = {
    "AAPL": 180,
    "TSLA": 250,
    "GOOGL": 140,
    "AMZN": 130,
    "MSFT": 320
}
print("available stocks and prices:")
for stock,price in stock_prices.items():
  print(f"{stock} - ${price}")

total_invest=0

n = int(input("how many different stocks do you want to buy? ")) # Moved this line outside the loop
for i in range(n):
  stock_symbol=input("enter the symbol").upper()
  quantity = int(input("enter quantity: "))
  if stock_symbol in stock_prices:
    total = stock_prices[stock_symbol]* quantity
    total_invest+=total
    print(f"{stock_symbol}: {quantity} shares × ${stock_prices[stock_symbol]} = ${total}")
  else:
      print("Invalid stock name. Please try again.")

print(f"\nYour total investment value: ${total_invest}")

save = input("save the result to a file?(yes/no): ").lower()
if save=="yes":
  with open("stock_invest.txt","w") as file:
    file.write(F"total ivestment value: ${total_invest}\n")
  print("Investment summary saved to 'stock_invest.txt'")
```
